```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
```

```python
points = 100
a = 4.0
vec2 = signal.ricker(points, a)
```

```python
plt.plot(vec2)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
xtop = range(100)
ytop = [1 if 75 > z > 25 else 0 for z in x]
```

```python
plt.plot(xtop,ytop)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
fig, ax = plt.subplots(1,3,figsize=(12,3))

ax[1].plot(vec2)
ax[1].set_xticks([])
ax[1].set_yticks([])
ax[1].set_title('Mexican Hat')

ax[0].plot(x, stats.norm.pdf(x, mu, sigma))
ax[0].set_xticks([])
ax[0].set_yticks([])
ax[0].set_title('Gaussian')

ax[2].plot(xtop,ytop)
ax[2].set_xticks([])
ax[2].set_yticks([])
ax[2].set_title('Top Hat')

plt.tight_layout()
plt.savefig('/home/mj1e16/THESIS/figures/filterShapes,png',dpi=100)
plt.show()
```
