```python
from astropy.table import Table, vstack
from astropy.io import ascii
from astropy.io import fits

import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.cm as cm

import keplerlcparams as lcp
import keplerlcparamsSN as lcpSN
import keplerlcparamsCV as lcpcv

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import AxesGrid
import matplotlib.colors as colors
```

```python
font = {'family' : 'serif',
        'weight' : 'normal',
        'size'   : 14}
plt.rc('font', **font)
plt.rc('axes',linewidth=1)
```

```python
def truncate_colormap(cmap, minval=0.0, maxval=1.0, n=100):
    new_cmap = colors.LinearSegmentedColormap.from_list(
        'trunc({n},{a:.2f},{b:.2f})'.format(n=cmap.name, a=minval, b=maxval),
        cmap(np.linspace(minval, maxval, n)))
    return new_cmap
```

```python
newplasma = truncate_colormap(cm.plasma_r,0.3,1.)
```

```python
origTabSN = ascii.read('/home/mj1e16/outTablesImproved/final/matchSNabove10.csv')
```

```python
for x in range(len(lcpSN.fullExt)):
    uni = list(set(lcpSN.fullExt[x]))
    pos = [float(uni.index(y))+10. for y in lcpSN.fullExt[x]]
    #pos.append(1.)
    if len(pos) > 1:
        plt.scatter(lcpSN.datesfull[x],lcpSN.mag[x],c=pos,cmap=newplasma,marker='D')
        plt.xlabel('Date (MJD)')
        plt.ylabel('Magnitude (mags)')
        plt.gca().invert_yaxis()
        plt.savefig('/home/mj1e16/thesisFigures/lcNewSN_{}_{}_{}.png'.format(origTabSN['RA'][x], origTabSN['DEC'][x],origTabSN['IMAGE_NAME'][x]))
        plt.show()
    #print(uni)
```
