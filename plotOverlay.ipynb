```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

import astropy
from astropy.coordinates import SkyCoord
from astropy import units as u

from astroquery.vizier import Vizier
```

```python
#catalogue = "V/106/lmxbcat" # lmxb
#catalogue = "J/A+A/572/A89 " # yso
#catalogue = "V/123A/cv" # cv
#catalogue = "J/A+A/435/521" # agn
catalogue = "V/115/algols" # eclips Bs
```

```python
Vizier.ROW_LIMIT = -1
lmxbs = Vizier.get_catalogs(catalogue)
```

```python
lmxbs[0]
```

```python
ra = []
dec = []
for x in range(len(lmxbs[0])):
    ra.append(lmxbs[0][x][-2])
    dec.append(lmxbs[0][x][-1])
```

```python
raDeg = []
decDeg = []
for x in range(len(ra)):
    c = SkyCoord(ra[x]+' '+dec[x], unit=(u.hourangle, u.deg))
    raDeg.append(c.ra.degree)
    decDeg.append(c.dec.degree)
```

```python
img=mpimg.imread('kplrffi.png')
fig,ax = plt.subplots(figsize=(10,10))
#plt.plot([0,5,10],[0,5,10])
imgplot = plt.imshow(img, interpolation="bicubic",extent=[302.45,279,36,53])
plt.plot(ra,dec,'ko')
plt.show()
```
