```python
original = np.logspace(1,5)
#original = np.linspace(1,581)
mag3flux = [x*100**(-0.6) for x in original]
sn10flux = [10*(110**0.5)*(x**0.5) for x in original]

plt.plot(original,mag3flux,label='3 Mag increase')
plt.plot(original,sn10flux,label='10 SN ratio')
plt.xlabel('Source Flux')
plt.ylabel('Difference Image Flux')
plt.legend()
plt.show()
```

```python
original = range(1000,10**6)
snflux = [10*(x**0.5) for x in original]
magdiff = [2.5 * np.log10(o/(o-d)) for o,d in zip(original,snflux)]
magflux = [0.06*x for x in original]
snratio = [x/(o-x)**0.5 for o,x in zip(original,magflux)]
sncut = [10 for x in original]
magcut = [3 for x in original]
plt.plot(original,snratio,label='sn')
plt.plot(original,magdiff,label='mag')
plt.plot(original,sncut,label='sncut')
plt.plot(original,magcut,label='magcut')
plt.legend()
plt.show()

```

```python
original = range(6000,10**6)
difference = [5999 for x in original]
magdif = [2.5 * np.log10(o/(o-d)) for o,d in zip(original,difference)]
snratio = [d/(o**0.5) for d,o in zip(difference,original)]
sncut = [10 for x in original]
magcut = [3 for x in original]
plt.plot(original,magdif,label='magnituide')
plt.plot(original,snratio,label='sn')
plt.plot(original,sncut,label='sncut')
plt.plot(original,magcut,label='magcut')
plt.ylim([0,15])
plt.legend()
plt.show()
```
