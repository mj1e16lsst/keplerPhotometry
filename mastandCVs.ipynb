```python
from astroquery.mast import Observations
import pandas as pd
```

```python
objects = ['V0344 Lyr','MASTER OT J185838.73+460207.2','MASTER OT J191156.83+403101.6','ASASSN-15mt','V0585 Lyr','V1504 Cyg']
```

```python
obsTableList = []
for x in objects:
    obs_table = Observations.query_criteria(objectname=x,radius="2 arcsec",filters=["Kepler"])
    obsTableList.append(obs_table)
    df = obs_table.to_pandas()
    df = df[df['obs_collection']!='KeplerFFI']
    print(len(df))
```

```python
obs_table = Observations.query_criteria(objectname='V0585 Lyr',radius="0 arcsec",filters=["Kepler"])
df = obs_table.to_pandas()
df = df[df['obs_collection']!='KeplerFFI']
```
